```python
NOW = dt.datetime(2018,1,28)
df['TransactionTime'] = pd.to_datetime(df['TransactionTime'])
```

```python
rfmTable = df.groupby('UserId').agg({'TransactionTime': lambda x: (NOW - x.max()).days, 'TransactionId': lambda x: len(x), 'TotalAmount': lambda x: x.sum()})
rfmTable['TransactionTime'] = rfmTable['TransactionTime'].astype(int)
rfmTable.rename(columns={'TransactionId': 'frequency', 
                         'TotalAmount': 'monetary_value',
                         'TransactionTime': 'recency'}, inplace=True)
```

```python
rfmTable.head(5)
```

```python
quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
```

```python
segmented_rfm = rfmTable

def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
segmented_rfm['r_quartile'] = segmented_rfm['recency'].apply(RScore, args=('recency',quantiles,))
segmented_rfm['f_quartile'] = segmented_rfm['frequency'].apply(FMScore, args=('frequency',quantiles,))
segmented_rfm['m_quartile'] = segmented_rfm['monetary_value'].apply(FMScore, args=('monetary_value',quantiles,))
segmented_rfm.head()
```

```python
segmented_rfm['RFMScore'] = segmented_rfm.r_quartile.map(str) + segmented_rfm.f_quartile.map(str) + segmented_rfm.m_quartile.map(str)
segmented_rfm.head()
```
